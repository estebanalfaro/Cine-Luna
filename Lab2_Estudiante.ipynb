### 4. Prompt CTRE

Copiar el prompt generado siguiente el método visto en clase. 

Respuesta :  Se me ha solicitado crear un programa para la venta de tiquetes online del "Cine Luna".

Tarea: Debes crear el codigo para una aplicación Web con archivos separados, fácil de entender con los siguiente componentes y requerimientos, la interfaz debe ser modernizada y que el cliente pueda ver los precios de los asientos para que pueda escoger que tipo de asientos quiere y el IVA de 13% para que sea lo mas claro posible.    

1. Configuración de la sala:

- La sala contiene 5 filas: A, B, C, D, E.
- Cada fila tiene 8 asientos numerados del 1 al 8.
- Total de asientos: 40.
- La fila C corresponde a asientos premium.
- Los asientos A1 y A2 son para personas con discapacidad ( ponlos en otro color, sin diferencia de precio).

2. Precios base:

- Asiento estándar: ₡4,500.
- Asiento premium (fila C): ₡6,500.

3. Recargos y descuentos: 

- Películas en formato IMAX tienen un recargo del 40% sobre el precio base.
- Descuentos disponibles:
- Estudiante: 20%.
- Adulto mayor: 25%.
- Solo se puede aplicar un descuento por tiquete.
- En caso de múltiples condiciones, se aplicará el mayor descuento.

4 . Validación de descuentos:

- El sistema permite al usuario seleccionar el tipo de cliente (normal, estudiante o adulto mayor).
- El sistema no valida automáticamente si el usuario cumple con la condición seleccionada.
- La verificación del descuento se realiza de forma externa al sistema (por ejemplo, en el punto de ingreso al cine).

5. Impuestos:

- Se debe aplicar un IVA del 13% sobre el total final de la compra.

6. Manejo de asientos:

- Los siguientes asientos no están disponibles A5, B3, B4, C4, C5, D7, E1, E2.
- Los asientos seleccionados deben cambiar su estado a “reservados”.

7. Carrito de compra:

- El sistema debe permitir seleccionar hasta un máximo de 6 tiquetes por compra.
- Al seleccionar el primer asiento, se inicia un temporizador de 5 minutos, debes mostrarlo en minutos y segundos.
- El temporizador no se reinicia al agregar más asientos.
- Si el tiempo expira, los asientos seleccionados deben liberarse automáticamente.

8. Facturación:

Al confirmar la compra, el sistema debe generar una factura bien desglosada separada en columnas, ( precio base + precio con descuento si aplicó + IVA + recargo por IMAX) con lo que se solicita abajo:

- Lista de asientos seleccionados.
- Tipo de asiento (estándar o premium).
- Tipo de cliente (normal, estudiante, adulto mayor).
- Precio por tiquete.
- Monto del IVA.
- Total final.
- Fecha y hora de emisión.
